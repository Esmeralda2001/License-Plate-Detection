# Measuring speed
The following code is to measure the average speed of the detection model. The speed will vary based on your system's hardware. 

```python
average = []

path = r"C:/Users/midni/Desktop/School/INNO/opencv-anpr/license_plates/group3/"
counter = 0
files = os.listdir(path)

for pic in sorted(files):
    filename = os.fsdecode(pic)
    pic_path = path + filename
    start = time.time()
    plates = Detection.detect_and_return_plate(pic_path, model, cfg)
    average.append(time.time()-start)
```

```python
print("Total speed:", sum(average)/len(average))
```
